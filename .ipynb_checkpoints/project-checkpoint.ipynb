In this project, we will perform basic Exploratory Data Analysis (EDA) on the **Kyphosis disease Dataset**. 

Kyphosis is an abnormally excessive convex curvature of the spine. 


<img src="https://my.clevelandclinic.org/-/scassets/images/org/health/articles/17671-kyphosis" width="400">

Dataset contains 81 rows and 4 columns representing data on children who have had corrective spinal surgery. 
- **INPUTS:**
   1. Age: in month
  2.  Number: the number of vertebrae involve
  3.  Start: the number of the first (topmost) vertebra operated on
- **OUTPUTS:** Kyphosis which represents a factor with levels absent present indicating if a kyphosis (a type of deformation) was present after the operation.n.

## 4. List the average, minimum and maximum age (in years)

```python
def data_fun(row):
    age_year = row/12

    return age_year
```

```python
df["Age_Year"] = df["Age"].apply(data_fun)
df
```

```python
df["Age_Year"].max()
```

## 7. Scale the raw Age column (in months) using both standardization and Normalization. Perform a sanity check.   

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
min_max_scaler = MinMaxScaler()

data_norm = min_max_scaler.fit_transform(df["Age"].values.reshape(-1,1))
data_norm[:10]
```

```python
sns.histplot(pd.DataFrame(data_norm), kde = True)
```

```python
standard_scaler = StandardScaler()

data_stand = standard_scaler.fit_transform(df["Age"].values.reshape(-1,1))
data_stand[:10]
```

```python
sns.histplot(pd.DataFrame(data_stand), kde = True)
```

```python
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")

df_norm = pd.DataFrame(data_norm, columns=["value"])
df_stand = pd.DataFrame(data_stand, columns=["value"])
sns.histplot(df_norm["value"], kde = True, alpha = 0.7, color = "red", label = "Normalized")
sns.histplot(df_stand["value"], kde = True, alpha = 0.5, color = "blue", label = "Standardized")


plt.legend() 
plt.show()
```
